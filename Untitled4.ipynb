```python
import numpy as np
import matplotlib.pyplot as plt
v = 100 * 1000 / 3600 # m/s
g = 9.81 # m/s^2
theta = np.radians(45) # radians
t = np.linspace(0, 2*(v*np.sin(theta))/g, 100)
def traj():
  x = (v * np.cos(theta))*t
  y = (v * np.sin(theta))*t - (0.5 * g * t**2)
  return x, y
plt.plot(traj()[0], traj()[1])
plt.xlabel('x (m)')
plt.ylabel('y (m)')
plt.title('Projectile Motion')
plt.show()

```
